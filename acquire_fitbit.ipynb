```python
def get_fitbit():
    chunks = []
    path = '~/Documents/Coding/Codeup/Coursework/ds-methodologies-exercises/time_series/fitbit/'
    for i, element in enumerate(files):
        i = pd.read_csv(path+element, skiprows = 35, nrows=33, engine = 'python')
        chunks.append(i)
#         print(chunks)
    return chunks
```
